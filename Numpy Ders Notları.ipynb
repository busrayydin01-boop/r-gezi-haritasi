## np.array().concatenate()

```python
a = np.array([1,2,3])
b = np.array([3,4,5])
c = np.array([5,6,7])
np.concatenate([a,b,c]) # a,b ve c arraylerini birleştirir.
```

```python
a2 = np.array([[1,2,3], [4,5,6]])
b2 = np.array([[1,1,1], [2,2,2]])
np.concatenate([a2,b2]) # a2 ve b2 iki boyutlu array'lerini birleştirir. #yatay olarak birleştirir.
```

```python
np.concatenate([a2,b2], axis=1) # axis birleştirmede baz alınacak ekseni ifade eder. normalde 0'dır.#dikey olarak birleştirdi.
```

```python
a=np.arange(1,21).reshape(4,5)
b= np.arange(21,41).reshape(4,5)

x=np.concatenate([a,b])

a=np.arange(1,21).reshape(4,5)
b= np.arange(21,41).reshape(4,5)

x=np.concatenate([a,b], axis=1) #dikey birleştircek.
```

## np.array().vstack()

```python
a = np.array([1,2,3])
b = np.array([[3,4,5],
              [5,6,7]])

np.vstack([a,b]) # sütun sayısı aynı, satır sayısı farklı olan
# array'leri DİKEY (vertical) olarak birleştirir.

# SÜTUN SAYISI AYNI OLMAZSA DİKEY OLARAK BİRLEŞTİREMEYİZ!
```

## np.array().hstack()

```python
a = np.array([1,2,3])
b = np.array([[3,4,5],
              [5,6,7]])

np.hstack([a,b]) # satır sayısı aynı sütun sayısı farklı olan
# array'leri YATAY (horizontal) olarak birleştirir.

# Aynı örneği kullanırsak, burada hata meydana gelir.
# Çünkü satır sayısı aynı değil!
```

```python
a = np.array([[1,2,3],
              [0,0,0]])
b = np.array([[4,5,6,7],
              [1,1,1,1]])

np.hstack([a,b]) # satır sayısı aynı sütun sayısı farklı olan
# array'leri YATAY (horizontal) olarak birleştirir.

# SATIR SAYISI AYNI OLMAZSA YATAY OLARAK BİRLEŞTİREMEYİZ!
```
